```python
#truth value (truthy or falsy), not the meaning of the content.
# Empty string "" → False
# Any non-empty string → True

print(bool("False"))
print(bool(0), bool(-1))
# 0 → False
# Any other number → True
print(bool([]), bool([0]), bool([False]), bool([None]))
```
